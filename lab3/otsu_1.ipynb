```python
img = cv2.imread("./test/test_img5.jpg", cv2.IMREAD_GRAYSCALE)
blur = cv2.GaussianBlur(img, (15, 15), 0)


thresh, binaryImg = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)


kernel = np.ones((10, 10), np.uint8)
closed_img = cv2.morphologyEx(binaryImg, cv2.MORPH_CLOSE, kernel, iterations = 1)

plt.imshow(binaryImg, cmap = 'gray')
```
